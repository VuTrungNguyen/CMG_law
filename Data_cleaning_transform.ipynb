```python
# normalize the table on Civil Code ID
df_cases["Civil Code ID"] = df_cases["Civil Code ID"].str.split(",")
df_cases["Civil Code Chinese Number"] = df_cases["Civil Code Chinese Number"].str.split(",")
normalized_df_cases = df_cases.explode(["Civil Code ID", "Civil Code Chinese Number"], ignore_index=True)
```

```python
normalized_df_cases.head()
```

```python
# save normalized_df_cases to xlsx
normalized_df_cases.to_excel('normalized_cases.xlsx', index=False)
```

```python
# Save the normalized df
import csv
normalized_df_cases.to_csv(
    'normalized_cases.csv',  # File name
    index=False,             # Exclude row indices
    quotechar='"',           # Enclose strings in double quotes
    quoting=csv.QUOTE_MINIMAL,  # Only quote fields containing special characters
    escapechar='\\',          # Escape special characters
    encoding='utf-8'          # Export the file in UTF-8 encoding
)
```

```python
# load normlized csv
csv_n_df = pd.read_csv("normalized_cases.csv")
csv_n_df.dropna()
```

```python
import csv
csv_n_df.to_csv(
    'normalized_cases.csv',  # File name
    index=False,             # Exclude row indices
    quotechar='"',           # Enclose strings in double quotes
    quoting=csv.QUOTE_MINIMAL,  # Only quote fields containing special characters
    escapechar='\\',          # Escape special characters
    encoding='utf-8'          # Export the file in UTF-8 encoding
)
```

```python
# Update the 'Title' and 'Court Finding' columns to ensure text with commas is quoted
for column in normalized_df_cases.columns:
    normalized_df_cases[column] = normalized_df_cases[column].apply(lambda x: f'"{x}"' if isinstance(x, str) and ',' in x else x)

# Save the modified DataFrame back to a CSV file
normalized_df_cases.to_csv('quoted_n_cases.csv', index=False)  # quoting=1 corresponds to QUOTE_MINIMAL
```
